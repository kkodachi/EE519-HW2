## Recordings

Here are helper functions to record and save my clips from previous assignments.

```python
import os
import csv
import numpy as np
from scipy.io import wavfile
from IPython.display import Audio

def write_csv(sentence_id: str, text: str, is_grammatical: int,
                    save_path: str = "project/sentences/ground_truth.csv"):
    """
    Records a sentence into a CSV file.

    Args:
        sentence_id (str): Sentence ID (e.g., 's01', 's02', ..., 's10')
        text (str): Ground truth sentence
        is_grammatical (int): 1 if grammatical, 0 if ungrammatical
        save_path (str): Path to save the CSV file
    """

    # Validate inputs
    if not sentence_id.startswith("s") or not sentence_id[1:].isdigit():
        raise ValueError("sentence_id must be in format 's01', 's02', etc.")

    if is_grammatical not in (0, 1):
        raise ValueError("is_grammatical must be 0 or 1")

    # Ensure directory exists
    os.makedirs(os.path.dirname(save_path), exist_ok=True)

    file_exists = os.path.isfile(save_path)

    # Write to CSV
    with open(save_path, mode="a", newline="", encoding="utf-8") as csvfile:
        writer = csv.writer(csvfile)

        # Write header if file does not exist
        if not file_exists:
            writer.writerow(["sentence_id", "text", "is_grammatical"])

        writer.writerow([sentence_id, text, is_grammatical])

    print(f"Saved sentence {sentence_id} to {save_path}")

SOUNDDEVICE_OK = False
try:
    import sounddevice as sd
    SOUNDDEVICE_OK = True
except Exception:
    SOUNDDEVICE_OK = False

def _to_float_mono(y):
    """Convert audio array to float32 mono in [-1, 1]."""
    y = np.asarray(y)
    if y.ndim == 2:
        y = y.mean(axis=1)
    if np.issubdtype(y.dtype, np.integer):
        y = y.astype(np.float32) / (np.iinfo(y.dtype).max + 1e-12)
    else:
        y = y.astype(np.float32)
    mx = np.max(np.abs(y)) + 1e-12
    if mx > 1.0:
        y = y / mx
    return y

def rms(y):
    y = np.asarray(y, dtype=np.float32)
    return float(np.sqrt(np.mean(y**2) + 1e-12))

def record_audio(seconds=5.0, fs=16000, channels=1):
    if not SOUNDDEVICE_OK:
        raise RuntimeError("sounddevice not available. Upload WAVs or install sounddevice locally.")
    print(f"Recording {seconds:.1f}s at {fs} Hz... (speak now)")
    y = sd.rec(int(seconds*fs), samplerate=fs, channels=channels, dtype='float32')
    sd.wait()
    y = _to_float_mono(y)
    print("Done. RMS:", rms(y))
    return fs, y

def save_wav(path, fs, y):
    """Save float audio in [-1,1] to 16-bit PCM WAV."""
    y16 = np.clip(y, -1.0, 1.0)
    y16 = (y16 * 32767).astype(np.int16)
    wavfile.write(str(path), fs, y16)

def play_audio(y, fs, label=None):
    if label:
        print(label)
    display(Audio(y, rate=fs))
```

### Sentence 1 Clean

```python
sentence_id, text, is_grammatical = "s01", "I go to the University of Southern California", 1
fs, y = record_audio()
play_audio(y, fs, label=f"Recorded: {sentence_id}")
save_wav(f"project/audio/clean/{sentence_id}.wav",fs,y)
write_csv(sentence_id,text,is_grammatical)
```
